### quick findings

**ibm**
- depth is **reduced** in both RB ↔ RR and CTM ↔ RR, while staying the same in CTM ↔ RB
- within RB ↔ RR, the **RB → RR** direction gives a slightly stronger reduction than the reverse 
- for CTM ↔ RR, both directions behave similarly
- **all reductions from 1-qubit gates** (no 2-qubit change)
- CTM ↔ RB only removes barriers, no structural change

**quantinuum**
- the same overall pattern appears  
- RB ↔ RR behaves symmetrically (no directional advantage)  
- in contrast, for CTM ↔ RR, the **RR → CTM** direction gives a slightly larger reduction than CTM → RR 
- no 2-qubit gate changes
